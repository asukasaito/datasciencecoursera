#### Data Preprocessing
- A significant portion of the 'location' data was missing or not standardized. For simplicity, these entries were imputed with the placeholder 'unknown'. Keywords were preserved as-is, given their potential predictive value.
- The 'text' column contained raw tweets with special characters, URLs, and stop words. These elements were removed, and the text was normalized and tokenized. This preprocessing is vital to ensure that the model learns from pertinent features.
- Global Vectors for Word Representation (GloVe) were utilized to embed the words into a meaningful vector space.

```python
from sklearn.impute import SimpleImputer
import re
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

# Impute missing values
imputer = SimpleImputer(strategy='constant', fill_value='unknown')
train_data[['keyword', 'location']] = imputer.fit_transform(train_data[['keyword', 'location']])
test_data[['keyword', 'location']] = imputer.fit_transform(test_data[['keyword', 'location']])

# Function to clean text data
def clean_text(text):
    text = text.lower()  # convert to lowercase
    text = re.sub(r'http\S+', '', text)  # remove URLs
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # remove special characters and numbers
    return text

# Apply text cleaning function
train_data['text'] = train_data['text'].apply(clean_text)
test_data['text'] = test_data['text'].apply(clean_text)

# Tokenization and Padding
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_data['text'])
sequences = tokenizer.texts_to_sequences(train_data['text'])

# Padding sequences to have the same length
max_seq_length = max([len(x) for x in sequences])  # You can also set a predefined max length
padded_sequences = pad_sequences(sequences, maxlen=max_seq_length)

# Now, padded_sequences is your input for the model
```

```python
from keras.preprocessing.text import Tokenizer

# Initialize the tokenizer with a max number of words
tokenizer = Tokenizer(num_words=10000)

# Fit the tokenizer on the training data
tokenizer.fit_on_texts(train_data['text'])

# Convert the text to sequences
train_sequences = tokenizer.texts_to_sequences(train_data['text'])
test_sequences = tokenizer.texts_to_sequences(test_data['text'])
```

```python
from keras.preprocessing.sequence import pad_sequences

# Find the length that covers 95% of the tweets
max_len = int(train_data['text'].apply(lambda x: len(x.split())).quantile(0.95))
max_len_test = int(test_data['text'].apply(lambda x: len(x.split())).quantile(0.95))

# Pad the sequences to have the same length
train_padded = pad_sequences(train_sequences, maxlen=max_len)
test_padded = pad_sequences(test_sequences, maxlen=max_len_test)
```

```python
od.download('https://www.kaggle.com/datasets/anindya2906/glove6b/data/glove.6B.100d.txt')
```

```python

# Create an index mapping words (from the embedding) to their vector representation
embedding_index = {}

with open('glove6b/glove.6B.100d.txt', 'r', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embedding_index[word] = coefs

# Prepare the embedding matrix
embedding_dim = 100
embedding_matrix = np.zeros((len(tokenizer.word_index) + 1, embedding_dim))

for word, i in tokenizer.word_index.items():
    embedding_vector = embedding_index.get(word)
    if embedding_vector is not None:
        # Words not found in the embedding index will be all zeros.
        embedding_matrix[i] = embedding_vector
```

```python
from keras import backend as K

def f1_score(y_true, y_pred):
    # Convert predictions to binary (0 or 1)
    y_pred_binary = K.round(y_pred)

    # True positives
    tp = K.sum(K.cast(y_true*y_pred_binary, 'float'), axis=0)
    # False positives
    fp = K.sum(K.cast((1-y_true)*y_pred_binary, 'float'), axis=0)
    # False negatives
    fn = K.sum(K.cast(y_true*(1-y_pred_binary), 'float'), axis=0)

    # Precision
    p = tp / (tp + fp + K.epsilon())
    # Recall
    r = tp / (tp + fn + K.epsilon())

    # Calculate f1_score
    f1 = 2*p*r / (p+r+K.epsilon())
    f1 = K.mean(f1)
    return f1

```

# Model Architecture:
Four types of models/hyper parameter pattens are examined.

**Simple RNN:** A single RNN layer captures the sequence pattern in tweet text, serving as a baseline for performance comparison.

**GRU RNN:** The GRU layer's ability to handle vanishing gradients makes it superior to Simple RNN for longer dependencies, which is common in language.

**RNN with GRU and Dropout:** Adding Dropout to the GRU model aims to prevent overfitting, ensuring the model generalizes better to unseen data.

**RNN with GRU (150 embedding_dim):** Increased embedding dimensions allow the model to encode more complex information but risk overfitting, which aim to balance with Dropout.

## Hyperparameter Tuning :
**Tuning:** Hyperparameters like the number of GRU units, the dropout rate, and the embedding dimension were adjusted. The GRU units were selected to balance the model's capacity without incurring too much computational cost. The dropout rate was chosen to prevent overfitting while allowing enough flexibility for the model to learn from the data.



```python
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, Dense

# Set the vocabulary size, embedding dimension, and maximum sequence length
vocab_size = 10000
embedding_dim = 50
max_length = max_len  # As calculated from the 95% quantile

# Build the model
simple_rnn_model = Sequential()
simple_rnn_model.add(Embedding(vocab_size, embedding_dim, input_length=max_length))
simple_rnn_model.add(SimpleRNN(64))
simple_rnn_model.add(Dense(1, activation='sigmoid'))

# Compile the model
simple_rnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', f1_score])

# Model summary
simple_rnn_model.summary()
```

```python
from keras.layers import GRU

# Build the model
gru_rnn_model = Sequential()
gru_rnn_model.add(Embedding(vocab_size, embedding_dim, input_length=max_length))
gru_rnn_model.add(GRU(64))  # GRU layer
gru_rnn_model.add(Dense(1, activation='sigmoid'))

# Compile the model
gru_rnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', f1_score])

# Model summary
gru_rnn_model.summary()
```

```python
# Train the simple RNN model
history_simple = simple_rnn_model.fit(train_padded, train_data['target'],
                               epochs=5,
                               validation_split=0.2,  # Use part of the training data for validation
                               verbose=1)
```

```python
from keras.layers import Dropout


model_with_dropout = Sequential()
model_with_dropout.add(Embedding(vocab_size, embedding_dim, input_length=max_length))
model_with_dropout.add(GRU(64, return_sequences=True))
model_with_dropout.add(Dropout(0.5))  # Add dropout with a 50% rate of dropout
model_with_dropout.add(GRU(64))  # Second GRU layer
model_with_dropout.add(Dropout(0.5))  # Add dropout with a 50% rate of dropout
model_with_dropout.add(Dense(1, activation='sigmoid'))

# Compile the model
model_with_dropout.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', f1_score])
```

```python
import numpy as np

# Function to generate predictions and save to CSV
def save_predictions_to_csv(model, test_padded, model_name, test_ids):
    # Generate predictions
    predictions = model.predict(test_padded)
    # Convert probabilities to binary predictions (0 or 1) based on a threshold of 0.5
    predictions_binary = (predictions > 0.5).astype(int)

    # Prepare the submission DataFrame
    submission = pd.DataFrame({'id': test_ids, 'target': predictions_binary.flatten()})

    # Save the DataFrame to a CSV file
    submission.to_csv(f'submission_{model_name}.csv', index=False)

test_ids = test_data['id']

# Save predictions for each model
save_predictions_to_csv(simple_rnn_model, test_padded, 'simple_rnn', test_ids)
save_predictions_to_csv(gru_rnn_model, test_padded, 'gru_rnn', test_ids)
save_predictions_to_csv(model_with_dropout, test_padded, 'dropout', test_ids)
save_predictions_to_csv(model_with_dropout_150, test_padded, 'dropout_150', test_ids)
```

- The baseline Simple RNN displayed a significant gap between training and validation F1 scores, indicating a strong tendency to overfit. - Introducing Dropout to the GRU RNN was intended to mitigate overfitting by randomly omitting neuron contributions during training. While the Kaggle F1 score was slightly lower than the baseline, it's notable that the validation F1 did not improve significantly, suggesting that the model might not have fully utilized its capacity.

- Enhancing the GRU RNN with a higher embedding dimension (150dim) aimed to improve the model's ability to capture more complex features. However, this did not translate into better performance, with the Kaggle F1 score matching the model with dropout and showing the highest overfitting potential given the training F1 score.

#### Analysis Conclusion:
The GRU-based models demonstrated potential but also highlighted the challenges in achieving the right balance between model complexity and generalization. The results emphasize the delicate nature of model tuning, particularly in the context of dropout and embedding sizes.
